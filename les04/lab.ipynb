6. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

```python
# Ваш код здесь
# Группируем по возрасту
young_group = df[df['Age'] < 30]
old_group = df[df['Age'] > 60]

young_survival_rate = young_group['Survived'].mean()
old_survival_rate = old_group['Survived'].mean()

young_survival_rate, old_survival_rate
```

```python
# Ваш код здесь
survival_rate_by_gender = df.groupby('Sex')['Survived'].mean()
survival_rate_by_gender
```

8.  Найдите самое популярное имя среди пассажиров Титаника мужского пола?

```python
# Ваш код здесь

male_passengers = df[df['Sex'] == 'male']

male_names = male_passengers['Name'].apply(lambda x: x.split(',')[1].split()[1] if len(x.split()) > 1 else x.split()[0])

most_common_name = male_names.mode().iloc[0]
print(f"The most common name among male passengers is: {most_common_name}")
```

### Дополнительные вопросы для защиты
1. Что такое медиана? Рассчитать медиану для предложенного ряда чисел.
Медиана — это значение, которое делит отсортированный набор данных на две равные части: половина значений будет меньше медианы, а половина — больше. Если количество элементов нечетное, медианой будет средний элемент. Если четное, то медиана — это среднее значение двух средних элементов.
2. Что такое дисперсия? Рассчитать дисперсию для предложенного ряда чисел.
Дисперсия измеряет, насколько сильно значения отклоняются от среднего значения. Она равна среднему квадрату отклонений каждого значения от среднего.
3. Что такое среднеквадратичное отклонение? Рассчитать среднеквадратичное отклонение для предложенного ряда чисел.
Среднеквадратичное отклонение — это квадратный корень из дисперсии. Оно дает представление о том, насколько сильно значения отклоняются от среднего.
4. Что такое Series и чем он отличается от DataFrame?
Series — это одномерный массив с метками (индексами), который может хранить данные различных типов (например, числа, строки). Series имеет индекс и значения.

DataFrame — это двумерная таблица данных, которая состоит из строк и столбцов. Каждый столбец может быть Series, и все столбцы могут иметь разные типы данных.
5. Можно ли изменить индекс в Series после создания объекта?
Да, индекс в Series можно изменить с помощью атрибута .index.
6. Можно ли выполнить операцию сложения над двумя объектами Series, если они имеют:
Одинаковые типы данных, одинаковые размеры и одинаковые индексы? Да, можно.
Одинаковые типы данных, одинаковые размеры, но разные индексы? Нет, индексы должны совпадать, чтобы операции прошли без ошибок.
Одинаковые типы данных, разные размеры и одинаковые индексы? Нет, размеры должны совпадать.
Разные типы данных, одинаковые размеры и одинаковые индексы? Нет, операцию можно выполнить только если типы данных совместимы.
Разные типы данных, одинаковые размеры и разные индексы? Нет, типы данных должны быть совместимы, и индексы тоже.
Разные типы данных, разные размеры и разные индексы? Нет, операция сложения не будет выполнена.
7. Что такое DataFrame и чем он отличается от Series?
DataFrame — это двумерная структура данных с метками для строк и столбцов, где каждый столбец представляет собой отдельную серию данных. Он отличается от Series тем, что может содержать более одного столбца с данными, тогда как Series — это одномерный объект.

9. Чем отличается метод iloc от метода loc?
iloc — используется для выбора данных по целочисловым позициям (индексам).
loc — используется для выбора данных по меткам индекса или имени столбца.

11. Требуются ли установка дополнительных библиотек для загрузки с помощью pandas файлов в формате CSV? Excel (.xls)? Json?
Для загрузки CSV и JSON файлов не требуется установка дополнительных библиотек. pandas поддерживает их из коробки.
